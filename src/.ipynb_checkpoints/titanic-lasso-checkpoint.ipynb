**Logistic Regression**

```python
#Defines the formula for logistic regression and creates a model with a summary
formula_log2 = 'survived ~ pclass + alone'
log_model_2 = smf.logit(formula_log2, data=titanic_cleaned_class).fit()
pickle.dump(log_model_2, open('../results/log_model_2.pkl', 'wb'))
log_model_2.summary()
```

```python
#Creates a contingency table of the probabilities and the observed values.
titanic_cleaned['predictions'] = log_model_2.predict()
titanic_cleaned.loc[titanic_cleaned['predictions'] >= .50, 'Predicted'] = 1
titanic_cleaned.loc[titanic_cleaned['predictions'] < .50, 'Predicted'] = 0
data_crosstab = pd.crosstab(titanic_cleaned['Predicted'],
                            titanic_cleaned['survived'], 
                               margins = True)

print(data_crosstab)
```

```python
#Creates the 'classification report'
print(classification_report((titanic_cleaned['survived']), titanic_cleaned['Predicted'].astype(float)))
```

```python
#Produce the 'odds ratio' for the model to be able to interpret the model coefficients
params = log_model_2.params
conf = log_model_2.conf_int()
conf['Odds Ratio'] = params
conf.columns = ['5%', '95%', 'Odds Ratio']
print(np.exp(conf))
```

```python
#Check linearity by examining the 'fitted values'
sns.regplot(x = log_model_2.fittedvalues, y = log_model_2.resid_dev)
plt.xlabel('Fitted values')
plt.ylabel('Deviance residuals')
plt.title('Linearity of independent variables and log-odds')
plt.savefig(fname = '../figures/Linearity of independent variables and log-odds.png', format = 'png', dpi = 300, bbox_inches='tight')
plt.show()
```

```python
#harvey collier test for linearity 
sms.linear_harvey_collier(log_model_2)
```

```python
#create new column with resid_dev values
titanic_cleaned_class['residuals'] = log_model_2.resid_pearson

#filters column (resid_dev) by values and selects rows that match criteria
extreme_values_negative = titanic_cleaned_class.loc[(titanic_cleaned_class['residuals'] <= -3)]
extreme_values_positive = titanic_cleaned_class.loc[(titanic_cleaned_class['residuals'] >= 3)]

#shows all values that are greater than 3 within standard_resid column
extreme_values_negative
```
